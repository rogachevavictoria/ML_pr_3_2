```python
import numpy as np
import pandas as pd
import os

from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense
from keras.models import Sequential
```

```python
def create_model():
  model= keras.Sequential()
  model.add(layers.Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
  model.add(layers.Dense(8,kernel_initializer='uniform',activation='relu'))
  model.add(layers.Dense(1,kernel_initializer='uniform',activation='sigmoid'))
  
  model.compile(  optimizer = 'adam', loss = 'BinaryCrossentropy', metrics = ['accuracy'])
  return model
```

```python
dataset = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/data/diabetes.csv',delimiter=',',skiprows=1)
dataset.shape
```

```python
seed = 7
np.random.seed(seed)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
model = KerasClassifier(build_fn=create_model, nb_epoch=150, batch_size=10, verbose=0)
```

```python
model.get_params()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
kfold = StratifiedKFold(n_splits = 10, shuffle=True, random_state=seed)
```

```python
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
results = cross_val_score(model, X, Y,cv=kfold)
```

```python
results
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def create_model2(optimizer = 'rmsprop', init='glorot_uniform'):
  model= keras.Sequential()
  model.add(layers.Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
  model.add(layers.Dense(8,kernel_initializer='uniform',activation='relu'))
  model.add(layers.Dense(1,kernel_initializer='uniform',activation='sigmoid'))
  
  model.compile(  optimizer = 'adam', loss = 'BinaryCrossentropy', metrics = ['accuracy'])
  return model
```

```python
model1 = KerasClassifier(build_fn=create_model2,verbose=0)
```

```python
optimizers = ['rmsprop','adam']
init = ['glorot_uniform','normal','uniform']
epochs = np.array([50,100,150])
batches = np.array([5,10,20])
param_grid = dict(optimizer=optimizers,nb_epoch=epochs,batch_size=batches,init=init)
grid = GridSearchCV(estimator=model1, param_grid=param_grid)
```

```python
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
grid_result = grid.fit(X,Y)
```

```python
print('Best: %f using %s' % (grid_result.best_score_,grid_result.best_params_))
```
